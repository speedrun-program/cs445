```python
import warnings

warnings.simplefilter(action = "ignore", category = FutureWarning)

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import seaborn as sns
```

```python
# the data shows student statistics from a math course in two different schools, Gabriel Pereira (GP) and Mousinho da Silveira (MS).

student_df = pd.read_csv('student-mat.csv', sep=';')
student_df.head(5)
```

```python
# a count plot of students interested in higher education

# what this shows: It shows how many students are interested in higher education, and how many aren't.
# highlight: It highlights that most students in both schools want higher education.
# what can be learned: Both schools are doing a good job keeping most of their students interested in going to college.

sns.countplot(x="higher", hue="school", data=student_df)
plt.title("count plot of students interested in higher education")
plt.xlabel("interested")
plt.ylabel("count")
plt.savefig("jack_count_plot.png")
plt.show()
```

```python
# a box plot of number of absenses by school

# what this shows: It shows the average and max number of absences per school.
# highlight: It highlights that students in GP tend to be absent more often than students in MS on average.
# what can be learned: MS students do a better job with their class attendance.

sns.boxplot(x="absences", hue="school", data=student_df)
plt.title("box plot of students interested in higher education")
plt.xlabel("school")
plt.ylabel("absences")
plt.savefig("jack_box_plot.png")
plt.show()
```

```python
# violin plots of student period 1, 2, and 3 grades vs whether or not they attended nursery school

# what this shows: It shows the grades for students in each period, how many students got that
#                  grade level, and how many which got that grade level did or didn't attend nursery school.
# highlight: It highlights that students who attended nursery school get slightly lower grades on average, but there's little difference.
# what can be learned: Nursery school doesn't tend to improve student grades later in life.

sns.violinplot(x="G1", hue="nursery", data=student_df, cut=0, split=True)
plt.title("violin plot of student 1st period grades vs nursery school attendance")
plt.xlabel("period 1 grade")
plt.ylabel("attended nursery school")
plt.xticks((0.0, 2.5, 5.0, 7.5, 10.0, 12.5, 15.0, 17.5, 20.0))
plt.savefig("jack_violin_plot_G1.png")
plt.show()

sns.violinplot(x="G2", hue="nursery", data=student_df, cut=0, split=True)
plt.title("violin plot of student 2nd period grades vs nursery school attendance")
plt.xlabel("period 2 grade")
plt.ylabel("attended nursery school")
plt.xticks((0.0, 2.5, 5.0, 7.5, 10.0, 12.5, 15.0, 17.5, 20.0))
plt.savefig("jack_violin_plot_G2.png")
plt.show()

sns.violinplot(x="G3", hue="nursery", data=student_df, cut=0, split=True)
plt.title("violin plot of student 3rd period grades vs nursery school attendance")
plt.xlabel("period 3 grade")
plt.ylabel("attended nursery school")
plt.xticks((0.0, 2.5, 5.0, 7.5, 10.0, 12.5, 15.0, 17.5, 20.0))
plt.savefig("jack_violin_plot_G3.png")
plt.show()
```

```python
# regplots plotting number of absences vs grade in periods 1, 2, and 3

# what this shows: It shows the correlation between absences and grades
# highlight: It highlights that there's surprisingly little correlation between absences and grades
# what can be learned: To improve student grades, it may be better to focus on things other than class attendance first

sns.regplot(x="absences", y="G1", data=student_df)
plt.title("regplot of absences vs 1st period grades")
plt.xlabel("absences")
plt.ylabel("1st period grades")
plt.savefig("jack_reg_plot_G1.png")
plt.show()

sns.regplot(x="absences", y="G2", data=student_df)
plt.title("regplot of absences vs 2nd period grades")
plt.xlabel("absences")
plt.ylabel("2nd period grades")
plt.savefig("jack_reg_plot_G2.png")
plt.show()

sns.regplot(x="absences", y="G3", data=student_df)
plt.title("regplot of absences vs 3rd period grades")
plt.xlabel("absences")
plt.ylabel("3rd period grades")
plt.savefig("jack_reg_plot_G3.png")
plt.show()
```

```python
# a histogram plot plotting student health (1 is bad, 5 is good)

# what this shows: The health levels of the students
# highlight: A lot of students, albeit a minority of them, report having either poor or very poor health
# what can be learned: There are a significant minority of students who have poor or very poor health.

sns.histplot(student_df["health"], discrete=True)
plt.title("histogram plot of student health")
plt.xlabel("health level (1 is bad, 5 is good)")
plt.ylabel("students with this health level")
plt.savefig("jack_histogram_plot.png")
plt.show()
```
